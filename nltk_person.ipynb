```python
import nltk

nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

paragraph = "I'm going through some things with my feelings and myself. \
I barely sleep and I do nothing but think about how I'm worthless and how I shouldn't be here.\
   I've never tried or contemplated suicide. I've always wanted to fix my issues, but I never get around to it.\
   How can I change my feeling of being worthless to everyone?"



sid = SentimentIntensityAnalyzer()
print('input: ' + paragraph)
ss = sid.polarity_scores(paragraph)

print(ss)
```

```python
import matplotlib.pyplot as plt

#시각화
labels = ['Negative','Neutral','Positive']
sizes = [ss['neg'],ss['neu'],ss['pos']]
colors = ['lightsteelblue','cornflowerblue','royalblue']


plt.figure(figsize=(10,10))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140)
plt.title("Result of Sentiment analysis")


plt.show()
```
